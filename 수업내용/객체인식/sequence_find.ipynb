```python
import json
from datetime import datetime
import os
from tqdm import tqdm

def jsonfind(json_file):
    find_file_path = []
    with open(json_file, 'r', encoding='utf-8') as f:
        opened_file = json.load(f)
    time = opened_file['time']
    time = datetime.strptime(time, "%H:%M:%S")
    fr_time = datetime.strptime("09:20:00", "%H:%M:%S")
    la_time = datetime.strptime("17:40:00", "%H:%M:%S")

    if time < fr_time or time > la_time:
        find_file_path.append(json_file)
    
    return find_file_path

def find_json_files(base_path):
    all_find_file_paths = []
    
    json_path_list = [f for f in os.listdir(base_path) if f.endswith('.json')]
    print(f"Found {len(json_path_list)} JSON files in directory {base_path}")
    
    for json_path in tqdm(json_path_list, desc="Processing files"):
        find_file_path = jsonfind(os.path.join(base_path, json_path))
        all_find_file_paths.extend(find_file_path)
    
    return all_find_file_paths

base_path = 'C:/Users/chldy/Downloads/resized_valid_annotations/resized_valid_annotations'
find_file_path = find_json_files(base_path)

# 결과 출력
print(f"Files found: {find_file_path}")
```
